### Read the data from database

```python
client = MongoClient()
client = MongoClient('localhost', 27017)
```

```python
db = client.airbnb
cursor = db.reviews.find()
```

```python
data = pd.DataFrame(list(cursor))
```

```python
data.head(10)
```

```python
from textblob import TextBlob
```

###### Create a new column for the english comments

```python
data["eng_comment"] = ""
```

```python
data.head(19110)
```

###### Translates comments if they are not in English. If the comment is already in English, add it into the list. If not, translate it and add it into the list. The length of the review has to be at least 3 words to be translatable. Hence, if the comment is fewer than 3 words, we will just add the existing comment into the list without translation. Also, if there is no review for a row (NA found), an empty string is added to the list
###### The printed output is the row number of the review
###### As translating the entire data takes hours, we only showed the translation of the first 10 rows

```python
lst = []
for i in range(0,10):
    str_comment = data.iloc[i].comments
    if pd.isnull(str_comment):
        #data.eng_comment.iloc[i] = ""
        lst.append([i])
        continue
    count = len(re.findall(r'\w+', str_comment))
    
    if count <3:
        #data.eng_comment.iloc[i] = str_comment
        lst.append([i,str_comment])
        continue
    if type(str_comment) == float:
        #data.eng_comment.iloc[i] = str_comment
        lst.append([i,str_comment])
        continue
    
    
    comment = TextBlob(data.iloc[i].comments)
    lang = comment.detect_language()
    print(i)
    if (lang == "en"):
        #data.eng_comment.iloc[i] = str_comment
        lst.append([i,str_comment])
        continue
    else:
        translation = comment.translate(from_lang=lang, to='en')
        #data.eng_comment.iloc[i] = str(translation)
        test = str(translation)
        lst.append([i,test])
    

test_df = pd.DataFrame()
a = test_df.append(lst)
```
